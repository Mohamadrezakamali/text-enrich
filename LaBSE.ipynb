```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import random as rd
import json
import numpy as np
import re
```

```python
sentence = "به ملازمان سلطان که رساند این دعا را که به شکر پادشاهی ز نظر مران گدا را ز رقیب دیوسیرت به خدای خود پناهم مگر آن شهاب ثاقب مددی دهد خدا را"
embedder = SentenceTransformer('sentence-transformers/LaBSE')
embeddings = embedder.encode(sentence)
print(embeddings)
```
