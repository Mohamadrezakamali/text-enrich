```python
def fast2tag (fast):
  columns = ["input_id","output_id","input_text","output_text","chapter","tag"]
  fast_tag_table = pd.DataFrame([], columns=columns)
  for i in range(len(fast)):
      oi = int(fast.iloc[i]["output_id"])
      if (oi <= 494 ):
        chapter_name = hafez.iloc[oi]["qazal_number"]
      elif (oi<= 2872):
        chapter_name = shams.iloc[oi-494-1]["qazal_number"]
      elif (oi>=5249 and oi<=5682):
        chapter_name = nahj.iloc[oi-5249]["hekmat_number"]
      else:
        chapter_name = ""
      fast_tag_table = fast_tag_table.append({'input_id':fast.iloc[i]["input_id"],'output_id':fast.iloc[i]["output_id"],'input_text':fast.iloc[i]["input_text"],'output_text':fast.iloc[i]["output_text"],'chapter':chapter_name,'tag':0}, ignore_index = True)
  return fast_tag_table
```
